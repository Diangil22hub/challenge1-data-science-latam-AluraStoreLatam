#1. Análisis de facturación



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas_data = [tienda1, tienda2, tienda3, tienda4]
ingresos = []

for i, tienda_df in enumerate(tiendas_data, start=1):
    tienda_df["Ingreso"] = tienda_df["Precio"]
    ingreso_total = tienda_df["Ingreso"].sum()
    ingresos.append(ingreso_total)
    print(f'El ingreso total de la tienda {i} es: {ingreso_total:,.2f}')

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(nombres_tiendas, ingresos, color='olive')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.xticks(rotation=15)
plt.tight_layout()

# Agregar etiquetas con los valores
for i, valor in enumerate(ingresos):
    plt.text(i, valor, f'{valor:,.2f}', ha='center', va='bottom')

plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

dfs = []

# Lee cada archivo y agregar una columna 'Tienda' para identificar su origen
for nombre_tienda, url in urls.items():
    df = pd.read_csv(url)
    df['Tienda'] = nombre_tienda
    dfs.append(df)

# Une todos los DataFrames en uno solo
df_total = pd.concat(dfs, ignore_index=True)

# Aseguramos el tipo de dato correcto
df_total['Categoría del Producto'] = df_total['Categoría del Producto'].astype(str)
df_total['Tienda'] = df_total['Tienda'].astype(str)

# Agrupamos los datos por tienda y categoría
cantidad_por_categoria = df_total.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad')

# Identificar la categoría MÁS popular por tienda
top_categorias = cantidad_por_categoria.loc[cantidad_por_categoria.groupby('Tienda')['Cantidad'].idxmax()]

# Añadimos columna "Destacado"
cantidad_por_categoria['Destacado'] = cantidad_por_categoria.apply(
    lambda row: 'Sí' if ((row['Tienda'], row['Categoría del Producto']) in zip(top_categorias['Tienda'], top_categorias['Categoría del Producto'])) else 'No',
    axis=1
)
print(cantidad_por_categoria)

# NUEVA PALETA DE COLORES - Tonos elegantes y profesionales
tiendas = cantidad_por_categoria['Tienda'].unique()
colores_base = ['#008B8B', '#FF8C00', '#FFF8DC', '#B22222']
paleta_completa = {}

# Para cada tienda, color base para destacado y tono claro para no destacado
for i, tienda in enumerate(tiendas):
    paleta_completa[(tienda, 'Sí')] = colores_base[i]  # Color intenso
    # Crear versión más clara del color para categorías no destacadas
    rgb = tuple(int(colores_base[i][j:j+2], 16)/255 for j in (1, 3, 5))
    rgb_claro = tuple([c + 0.5*(1-c) for c in rgb])
    paleta_completa[(tienda, 'No')] = rgb_claro

cantidad_por_categoria['Color'] = cantidad_por_categoria.apply(
    lambda row: paleta_completa[(row['Tienda'], row['Destacado'])],
    axis=1
)

# GRÁFICO DE BARRAS AGRUPADAS
plt.figure(figsize=(14, 7))

categorias = cantidad_por_categoria['Categoría del Producto'].unique()
ancho_barra = 0.2
desplazamientos = {
    'Tienda 1': -1.5 * ancho_barra,
    'Tienda 2': -0.5 * ancho_barra,
    'Tienda 3': 0.5 * ancho_barra,
    'Tienda 4': 1.5 * ancho_barra,
}

# Dibujar barras para cada tienda
for tienda in tiendas:
    datos_tienda = cantidad_por_categoria[cantidad_por_categoria['Tienda'] == tienda]
    posiciones = [i + desplazamientos[tienda] for i in range(len(categorias))]
    colores = datos_tienda['Color'].tolist()
    plt.bar(
        posiciones,
        datos_tienda['Cantidad'],
        width=ancho_barra,
        color=colores,
        edgecolor='black',
        linewidth=0.8,
        label=tienda
    )

# Formato del gráfico
plt.xticks(range(len(categorias)), categorias, rotation=30, ha='right')
plt.title('\nCategorías más vendidas por Tienda\n',
          fontsize=16, fontweight='normal')
plt.xlabel('Categoría del Producto', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.legend(title='Tienda', fontsize=10)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt

# PASO 1: Cargar los datos de cada tienda
tienda1 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv")
tienda2 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv")
tienda3 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv")
tienda4 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv")

# PASO 2: Calcular el costo de envío promedio
costo_envio_promedio_tienda1 = tienda1['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio tienda 1: {costo_envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio tienda 2: {costo_envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio tienda 3: {costo_envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio tienda 4: {costo_envio_promedio_tienda4:.2f}")

# PASO 3: Preparar datos para la gráfica
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [costo_envio_promedio_tienda1, costo_envio_promedio_tienda2,
                costo_envio_promedio_tienda3, costo_envio_promedio_tienda4]

# PASO 4: Crear la gráfica de barras
plt.figure(figsize=(8, 6))
plt.bar(tiendas, costos_envio, color='olive')
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envío Promedio')
plt.title('Costo de Envío Promedio por Tienda')

# Agregar los valores encima de cada barra
for i, v in enumerate(costos_envio):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom')

plt.show()
```

# Análisis de facturacion

Al observar los ingresos totales de las cuatro tiendas, puedo identificar que la Tienda 1 lidera el grupo con una cifra superior a los 11.500 millones, seguida muy de cerca por la Tienda 2 y la Tienda 3, que también superan los 10.900 millones. En el extremo opuesto se encuentra la Tienda 4, que presenta los ingresos más bajos del conjunto, con aproximadamente 10.380 millones. Aunque la diferencia no es abismal, esta posición rezagada en términos de facturación me hace pensar que la Tienda 4 podría estar enfrentando problemas para generar valor por unidad vendida o quizás no está capitalizando adecuadamente su volumen de ventas.

---

# Análisis de Ventas por Categoría

Al desglosar las ventas por categoría, encuentro patrones muy reveladores. La Tienda 1 domina claramente en las categorías de electrónicos y electrodoméstnicos, que son segmentos de alto valor y tradición comercial. Por su parte, la Tienda 3 también muestra un desempeño destacado en electrónicos, aunque con menor presencia en otras áreas. Sin embargo, lo que más me llama la atención es el comportamiento de la Tienda 4: sobresale en categorías como muebles, instrumentos musicales, deportes, juguetes y libros. Esta diversificación es en principio una fortaleza, pero al contrastarla con sus ingresos totales, me doy cuenta de que probablemente está compitiendo en segmentos de menor ticket promedio, lo que explica por qué sus ingresos no terminan de despegar a pesar de su amplitud de oferta.

---

# Análisis de Calificación Promedio

Al evaluar la percepción de los clientes a través de la calificación promedio, los datos me generan cierta preocupación. La Tienda 1 goza de una calificación de 4.0, lo que indica un alto nivel de satisfacción y consistencia en la experiencia de compra. En contraste, la Tienda 4 obtiene la calificación más baja, con apenas 2.5. Esta diferencia es significativa, especialmente si consideramos que la confianza del consumidor y la repetición de compra están fuertemente ligadas a la percepción de calidad. Una calificación tan baja en la Tienda 4 puede estar erosionando su base de clientes, independientemente de la cantidad de productos que logre vender.

---
# Análisis de Cantidad Vendida por Categoría

Al examinar la cantidad de unidades vendidas por categoría, confirmo una tendencia que ya había anticipado: la Tienda 4 es la que más vende en la mayoría de los rubros, liderando en electrodomésticos, muebles, instrumentos musicales, juguetes y especialmente en libros, donde duplica o triplica las ventas de las demás tiendas. Sin embargo, este liderazgo en volumen no se traduce en ingresos proporcionales, lo que refuerza mi impresión de que la Tienda 4 opera con productos de bajo precio unitario y posiblemente con márgenes reducidos. La Tienda 1 y la Tienda 3, en cambio, logran un mejor equilibrio entre volumen y valor, concentrándose en categorías más rentables.



---


# Análisis de Costo de Envío Promedio

El costo de envío es otro factor que analizo con detenimiento, ya que influye directamente en la decisión de compra del cliente. Aquí, la Tienda 4 presenta una ventaja competitiva notable, con el costo de envío más bajo del grupo. Esto podría explicar en parte su alto volumen de ventas en categorías como libros o juguetes, donde el precio de envío puede ser un factor decisivo. Sin embargo, esta ventaja logística no parece estar compensando sus debilidades en satisfacción al cliente, lo que sugiere que el bajo costo no es suficiente para retener a los compradores a largo plazo.

Después de analizar en profundidad todas las estadísticas, puedo afirmar que la Tienda 1 es la más sólida del grupo: combina altos ingresos, buena calificación, liderazgo en categorías estratégicas y un desempeño equilibrado en general.
